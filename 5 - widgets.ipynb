```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

**Widgets** are a way to make the notebooks interactive

# without widget 

We want to plot the **sin(alpha*x)** for alpha between 1 and 10

```python
alpha = 2

def sin_alpha(alpha, x):
    return np.sin(alpha * x)

x = np.arange(0,50)

plt.figure(0)
plt.plot(x, sin_alpha(alpha, x))
```

You need to change the alpha value manually and re run the cell to see the changes!

# with widgets 

```python
from IPython.html.widgets import interact
from IPython.html import widgets
```

```python
x = np.arange(0, 200)

def plot_sin_alpha(alpha):
    
    def sin_alpha(alpha, x):
        return np.sin(alpha * np.pi*2*x/200.)

    plt.figure()
    plt.plot(x, sin_alpha(alpha, x))

interact(plot_sin_alpha, alpha=widgets.IntSlider(min=0, max=10, vlaue=0))
```
