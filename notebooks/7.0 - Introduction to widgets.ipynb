# Problem 

## the old way 

We want to plot the **sin(alpha*x)** for alpha between 1 and 10, and x between 0 and 100

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
alpha = 1

def sin_alpha(alpha, x):
    return np.sin(alpha*x)

x = np.arange(0, 100)

f, ax = plt.subplots()
ax.plot(x, sin_alpha(alpha, x))
```

### Exercise

Play with alpha, value between 1 and 10, and notice how the plot change.


**NB**: Isn't that annoying?
There must be a better way!

## New way, using widgets

Widgets allow us to interact live with the plot by adding, in this case, a slider.

```python
from ipywidgets import widgets, interact
```

```python
x = np.arange(0, 100)

def plot_sin_alpha(alpha):
    
    def sin_alpha(alpha, x):
        return np.sin(alpha*x)

    plt.figure()
    plt.plot(x, sin_alpha(alpha, x))

interact(plot_sin_alpha, alpha=widgets.IntSlider(min=1, max=10, value=0))
```
