# Fits  ([documentation](http://docs.astropy.org/en/stable/io/fits/))

```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
```

## Load data 

Let's work with the following file

```python
fits_file = 'Data/Images/test_image.fits'
```

Always good to check if our file really exists

```python
import os
assert os.path.exists(fits_file)
```

We open the file and retrieve the data

```python
file_handle = fits.open(fits_file)
fits_array = file_handle[0].data
```

What is the shape of our array

```python
np.shape(fits_array)
```

Let's plot our image

```python
f1,ax1=plt.subplots(figsize=(5,5))
cax1=ax1.imshow(fits_array)
f1.colorbar(cax1)
```

Want a better plot?

```python
%matplotlib notebook
```

```python
f2,ax2=plt.subplots(figsize=(5,5))
cax2=ax2.imshow(fits_array)
f2.colorbar(cax2)
```
