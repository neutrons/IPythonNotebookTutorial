### Reading `csv` files as `numpy` arrays

For the function `np.genfromtxt` see: https://docs.scipy.org/doc/numpy/reference/generated/numpy.genfromtxt.html

````python
import os
import numpy as np
````

````python
# Use the same file from the previous notebook
file_path = os.path.join("..", "Data", "Glob", "f1.txt")
````

````python
# genfromtxt default
data = np.genfromtxt(
    file_path,
)
# Data is meanlingless!
data
````

````python
# genfromtxt delimiter and skip_header
data = np.genfromtxt(
    file_path,
    delimiter=",",
    skip_header=2,
)
# Data is 2D numpy array: 200 rows, 4 columns
data
````

````python
# What's its shape
data.shape
````

````python
# Let's get the first column
data[:,0]
````

````python
# We can do better! What about accessing the columns with its names?
# genfromtxt delimiter, skip_header and names
data = np.genfromtxt(
    file_path,
    delimiter=",",
    skip_header=2,
    names=['X', 'Y', 'E', 'DX'],
)
# Data is now a structure array
data
````

````python
# Attribute indexing: Structured arrays
data.dtype
````

Structured arrays are ndarrays whose datatype is a composition of simpler datatypes organized as a sequence of named fields.

````python
# X is a column in the file
data['X']
````

# Plotting

````python
# Choose your backend (the way the plots are rendered in a notebook)
%matplotlib notebook

# matplotlib.pyplot is a collection of command style functions that make matplotlib work like MATLAB
import matplotlib.pyplot as plt
````

````python
# Let's create shortcuts so it's easier to access the data
x = data['X']
y = data['Y']
xe = data['DX']
ye = data['E']
````

````python
# Simple case
fig, ax = plt.subplots()
ax.plot(x, y)
````

````python
# Let's label some stuff
# Simple Labels. Part 3 introduces axes manipulation
fig, ax = plt.subplots()
ax.set_title("I(q)")
ax.set_xlabel("$q (\AA^{-1})$")
ax.set_ylabel("$I (cm^{-1})$")
# Marker symbols: https://matplotlib.org/api/markers_api.html
ax.plot(x, y, marker='x', color='red', linestyle='')
````

````python
# Log x
fig, ax = plt.subplots()
ax.semilogx(x, y)
````

````python
# Log y
fig, ax = plt.subplots()
ax.semilogy(x, y)
````

````python
# Log X and Log Y
fig, ax = plt.subplots()
ax.loglog(x, y)
````

````python
## Error bars linear
fig, ax = plt.subplots()
ax.errorbar(x, y, xerr=xe, yerr=ye)
````

````python
## Error bars Lox Log y
fig, ax = plt.subplots()
ax.errorbar(x, y, xerr=xe, yerr=ye)
ax.set_xscale("log", nonposx='clip')
ax.set_yscale("log", nonposy='clip')
````

````python
# Same plot but let's prone some sata
fig, ax = plt.subplots()
ax.errorbar(
    x[2:-10],
    y[2:-10],
    xerr=xe[2:-10],
    yerr=ye[2:-10],
    # fmt = plot format string: '[color][marker][line]'
    # See Notes here: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html
    fmt='rx',
    ecolor='green',
    capthick=2,
)
ax.set_xscale("log", nonposx='clip')
ax.set_yscale("log", nonposy='clip')
````

````python
#let's suppose we want data for: Q > 0.02
data_x_subset_bool = x>0.02
data_x_subset_bool
````

````python
# Indexing data['X] with the condition created
x[data_x_subset_bool]
````

````python
# Length of the original X vs the prone X
# Note the print statetement order of values!
print("Original length = {1} :: Subset length = {0}.".format(
    len(x[data_x_subset_bool]),
    len(x),
))
````

````python
# Same plot but let's prone some sata
fig, ax = plt.subplots()
ax.errorbar(
    x[data_x_subset_bool],
    y[data_x_subset_bool],
    xerr=xe[data_x_subset_bool],
    yerr=ye[data_x_subset_bool],
    fmt='b.',
    ecolor='green',
)
ax.set_xscale("log", nonposx='clip')
ax.set_yscale("log", nonposy='clip')
````

## Create a new `CSV` file

- Imagine that you want to create a new plot using the same X coordinate and a new Y coordinate


````python
# Create a y_new exp(x * 10)
y_new = np.exp(data['X']*10)
y_new
````

````python
# Note the size of the X and Y arrays
x.size == y_new.size
````

````python
# Let's plot the new data just to see how it looks like
fig, ax = plt.subplots()
ax.plot(x, y_new)
````

### Adding the new array to the file

````python
# First import Recarray Helper Functions
from numpy.lib import recfunctions
````

````python
# Adding y_new array to the dataset we are working
data_new = recfunctions.append_fields(data, 'new', y_new)
````

````python
# Verify the type names now!
data_new.dtype.names
````

Let's add the new `data` to a file.

**Remember the original file header:**

```
# X , Y , E , DX
1
```

**We want this now:**

```
# X , Y , E , DX, new
1
```

````python
# Let's create the new header
# The first row:

header = data_new.dtype.names
# join: joins a list into string
header_str = " , ".join(header)
header_str
````

````python
# The 1st and the 2nd row:
header_str = "# " + header_str + "\n1"
print(header_str)
````

````python
# Let's save it in the directory tmp in our home directory 
import os

destination = os.path.expanduser("~/tmp")
print(destination)

os.makedirs(destination, exist_ok=True)

out_filepath = os.path.join(destination, "foo.txt")
print(out_filepath)
````

````python
# Use np.savetxt to save it

np.savetxt(
    out_filepath,
    data_new,
    delimiter=",",
    fmt='%.2f',
    header=header_str,
    comments='',
)

print("New dataset saved to:", out_filepath)
````

# Advanced Plotting

````python
# Let's create a single plot with two independent Y axis
# X vs Y and X vs new

fig, ax1 = plt.subplots()  # create figure and axes
ax1.plot(x, y, color='green')
ax1.set_ylabel('Old axis', color='green')

ax2 = ax1.twinx() # The magic is here
ax2.plot(x, y_new, color='red')
ax2.set_ylabel('new', color='red')

fig.suptitle('Advanced Plot', fontsize=20)
````

````python
# Let's create a plot, change the X ticks and create an annotation

fig, ax = plt.subplots()  # create figure and axes
ax.plot(x, y, color='red')
ax.set_xticks([0.001, 0.2, 0.4])
ax.set_xticklabels(['low', 'med', 'high'])
# https://matplotlib.org/users/annotations.html
ax.annotate(
    'This is the maximum!', 
    xy=(0.025, 1620), # Coordinate where the head is 
    xytext=(0.1, 1000), # Coordinate where the text starts
    arrowprops={'color':'b'}, # https://matplotlib.org/api/_as_gen/matplotlib.pyplot.annotate.html
)
fig.suptitle('Annotated Plot', fontsize=20, color='red')
````
