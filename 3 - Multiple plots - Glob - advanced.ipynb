# Python Glob: find files

glob is a powerful tool for finding files in the file system.
It is useful when you need to look for a list of files with names matching a pattern. 
For example, a list of filenames that all have a certain extension, prefix, or some common strings in the middle.

**Advanced:** The pattern rules for glob are not regular expressions. Instead, they follow standard Unix path expansion rules.

See https://docs.python.org/2/library/glob.html

# For Loop

```python
file_list = glob.glob('Data/Glob/*.txt')
for file in file_list:
    print(file)
```

```python
for i, file in enumerate(file_list):
    print(i, file)
```

```python
N = len(file_list)
for i in range(N):
    print(i, file_list[i])
```

\* zip

```python
N = len(file_list)
for i, file in zip(range(N), file_list):
    print(i, file)
```

## while loop

```python
N = len(file_list)
i = 0
while i<N:
    print(i, file_list[i])
    i+=1
```

# Plotting

find files

```python
file_list = sorted(glob.glob('Data/Glob/f*.txt'))
```

load data

```python
dataset = {}
for fn in file_list:
    dataset[fn] = np.genfromtxt(fn, delimiter=",", skip_header=2, names='X,Y,E,DX')
```

plot

```python
fig, ax = plt.subplots()
for fn in file_list:
    name = os.path.splitext(os.path.basename(fn))[0]
    data = dataset[fn]
    ax.semilogy(data['X'], data['Y'], label=name)
ax.legend()
ax.grid()
```

```python
markers = ('s', '+', 'p', 'o', '*', '^', 'v', '<', '>', '1') 

fig, ax = plt.subplots()
for fn, marker in zip(file_list, markers):
    name = os.path.splitext(os.path.basename(fn))[0]
    data = dataset[fn]
    ax.semilogy(data['X'], data['Y'], marker, markerfacecolor='white', label=name)
ax.legend()
ax.grid()
```

## Grid

```python
N = len(dataset)
ncols = 4
nrows = (N-1)//ncols+1
fig, axes = plt.subplots(nrows, ncols)
for irow in range(nrows):
    for icol in range(ncols):
        index = icol + irow * ncols
        if index >=N:break
        fn = file_list[index]
        name = os.path.splitext(os.path.basename(fn))[0]
        data = dataset[fn]
        ax1 = axes[irow][icol]
        ax1.semilogy(data['X'], data['Y'], label=name)
        ax1.legend()
        ax1.grid()
```

```python
fig = plt.figure()
N = len(dataset)
ncols = 4
nrows = (N-1)//ncols+1
for index, fn in enumerate(file_list):
    name = os.path.splitext(os.path.basename(fn))[0]
    data = dataset[fn]
    ax1 = fig.add_subplot(nrows, ncols, index+1)
    ax1.semilogy(data['X'], data['Y'], label=name)
    ax1.legend()
    ax1.grid()
    continue
```
